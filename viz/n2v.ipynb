```python
import torch
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
g = torch.load('../kg.pt', weights_only=False)
zs = torch.load('../n2v-10-64.pt', weights_only=False).detach()
```

```python
zs2 = PCA(n_components=2).fit_transform(zs)
```

```python
top_k = ["Kimsuky","APT29","Lazarus Group","MuddyWater","APT37","FIN7","Turla","APT28","APT41","Magic Hound","Mustang Panda","Andariel","TeamTNT","Transparent Tribe","OilRig","APT-C-36","Gamaredon Group","Sidewinder","TA505","APT38"]
groups = [g.node_names.index(apt) for apt in top_k]

zs = torch.load('../n2v-10-64.pt', weights_only=False).detach()
zs2 = PCA(n_components=2).fit_transform(zs)
for grp in groups:
    plt.text(zs2[grp, 0], zs2[grp,1], g.node_names[grp])

plt.axis('off')
plt.show()

zs = torch.load('../n2v-5-64.pt', weights_only=False).detach()
zs2 = PCA(n_components=2).fit_transform(zs)
for grp in groups:
    plt.text(zs2[grp, 0], zs2[grp,1], g.node_names[grp])

plt.axis('off')
plt.show()
```
