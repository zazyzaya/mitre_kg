```python
import torch
```

```python
# Taken from unseen pulse 67aa651978bbd23777b3db68 attributed to FIN7
ttps = ['UKRAINE', 'T1027', 'T1036', 'T1055', 'T1059', 'T1105', 'T1204', 'T1218', 'T1547', 'T1566', 'T1571', 'T1195', 'T1176', 'T1003', 'S0226', ]
zs = torch.load('../weights/n2v-10-64.pt', weights_only=False)
g = torch.load('../kg.pt', weights_only=False)

idxs = [g.nids.index(ttp) for ttp in ttps]
combined = zs[idxs] #.prod(dim=0, keepdim=True)

group_ids = (g.x[:, 9] == 1).nonzero().squeeze()
group_names = [g.node_names[grp] for grp in group_ids]
groups = zs[group_ids]

dist = torch.cdist(combined, groups)
dist = dist.mean(dim=0)
rank = dist.sort().indices

for r in rank:
    print(f'{group_names[r]}, {dist[r].item()}')
```
